```python
from pybiomart import Server

server = Server(host='http://www.ensembl.org')
dataset = server.marts['ENSEMBL_MART_ENSEMBL'].datasets['hsapiens_gene_ensembl']

gene_names = dataset.query(attributes=['ensembl_transcript_id_version', 'external_gene_name'])
gene_names.head(10)
```

```python
gene_names.columns = ["Name", "gene_name"]
merged = df.merge(gene_names, on="Name", how="left")
merged.head(10)
```

```python
top_named = merged.sort_values("TPM", ascending=False).head(20)
top_named[["gene_name", "TPM", "NumReads"]]
```
